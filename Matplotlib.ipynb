# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, color='blue', alpha=0.7)  # Optional: You can adjust the color and transparency

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show the plot
plt.grid(True)  # Optional: Adds a grid for better readability
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and axis with subplots
plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1, marker='o', color='blue', label='Line 1')
plt.title('Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y2, marker='s', color='orange', label='Line 2')
plt.title('Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.plot(x, y3, marker='^', color='green', label='Line 3')
plt.title('Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.plot(x, y4, marker='d', color='red', label='Line 4')
plt.title('Line 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Vertical Bar Plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.bar(company, profit, color='blue')
plt.title('Company Profit (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.grid(axis='y')

# Horizontal Bar Plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.barh(company, profit, color='green')
plt.title('Company Profit (Horizontal Bar Plot)')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.grid(axis='x')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It visually depicts the spread and skewness of the data, allowing for easy comparison between different datasets.

Why is a Box Plot Used?
Identifying Outliers: Box plots make it easy to see outliers in the data, which are points that fall significantly outside the range of the majority of the data.
Comparing Distributions: They allow for straightforward comparisons of different distributions, helping to identify differences in spread and central tendency.
Visualizing Spread: Box plots provide a visual summary of the central tendency (median) and variability (interquartile range) of the data.
Understanding Data Distribution: They offer insights into the shape of the distribution (e.g., skewness) without making any assumptions about the underlying distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Adding title and labels
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')

# Show the plot
plt.grid(axis='y')
plt.show()
```
