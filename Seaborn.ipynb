# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Here are five common plots that can be created using the Seaborn library, along with their uses:

1. **Scatter Plot (`sns.scatterplot`)**:
   - **Use**: Displays the relationship between two continuous variables. It is useful for visualizing patterns, clusters, or correlations in the data.
   - **Example**: Plotting height vs. weight to observe any relationship.

2. **Bar Plot (`sns.barplot`)**:
   - **Use**: Shows the relationship between a categorical variable and a continuous variable using rectangular bars. The height of the bar indicates the value.
   - **Example**: Visualizing average test scores of different categories (e.g., students from different schools).

3. **Box Plot (`sns.boxplot`)**:
   - **Use**: Summarizes the distribution of a continuous variable by displaying its quartiles and outliers. It is useful for detecting outliers and understanding the spread of the data.
   - **Example**: Showing the distribution of salaries in different departments.

4. **Heatmap (`sns.heatmap`)**:
   - **Use**: Visualizes data in a matrix format where individual values are represented by color intensity. Often used to show correlations between variables or in large datasets.
   - **Example**: Visualizing the correlation matrix of multiple variables.

5. **Violin Plot (`sns.violinplot`)**:
   - **Use**: Combines features of a box plot and a KDE (Kernel Density Estimate) plot. It shows the distribution, probability density, and spread of the data.
   - **Example**: Comparing the distribution of exam scores between male and female students.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for pclass vs. age
sns.boxplot(data=titanic, x="pclass", y="age", ax=axes[0])
axes[0].set_title('Pclass vs Age')

# Box plot for pclass vs. fare
sns.boxplot(data=titanic, x="pclass", y="fare", ax=axes[1])
axes[1].set_title('Pclass vs Fare')

# Display the plot
plt.tight_layout()
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram for price with hue for cut
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack", palette="Set2")

# Display the plot
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot with hue for species
sns.pairplot(data=iris, hue="species")

# Display the plot
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Create a pivot table to format the data for the heatmap
# Use keyword arguments for index, columns, and values
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")

# Display the plot
plt.show()
```
